```python
#Type of values
#Type of index
import pandas as pd
marks = pd.Series([78,85,90,66])
print(marks)
print("Type of values : ", marks.values)
print("Type of index: ", marks.index)


```

```python
#Type of values
#Type of index
import pandas as pd
marks = pd.Series([78,85,90,66],index=['a', 'b', 'c','d'])
print(marks)
print("Type of values : ", marks.values)
print("Type of index: ", marks.index)
```

```python
#Access Element
marks = pd.Series([78,85,90,66],index=['a', 'b', 'c','d'])
print("Access Element: ", marks["a"])
```

```python
#Explicit Index
marks1 = pd.Series([78,85,90,66],index=[1, 2, 3,4])
#print("Explicit Index included: ", marks1["a":"c"])
print(marks1.iloc[1:2])
print(marks1.loc[1:2])
```

```python
marks = pd.Series([78,85,90,66])
print("Implicit index included: ", marks[""])
```

```python
dict={"A":10,"B":20,"C":30}
marks=pd.Series(dict)
print("Series Object From Dictionary:", marks)
```

```python

data={"A":10,"B":20,"C":30}
marks=pd.Series(data)
print("Access with label-based slicing(explicit index):", marks["A":"C"])
print("Access with Position-based slicing:", marks[0:2])
```
